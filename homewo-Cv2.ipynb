```python
#引用需要的包
import numpy as np
from sklearn.datasets import load_boston
```

```python

'''
Step1:定义网络结构（指定输入层、隐藏层、输出层）

Step2：初始化模型参数

Step3：循环操作：
        1.执行前向传播
        2.计算损失函数
        3.执行后向传播
        4.权值更新
'''
##加载数据
data=load_boston()
X=data.data
Y=data.target
```

```python
X.shape
# print(X)
# 数据规范化处理
X = (X - np.mean(X_, axis=0)) / np.std(X, axis=0)
# print(X_)
```

```python
Y.shape
Y = Y.reshape(Y.shape[0],1)
```

```python
X_features=X.shape[1]
# print(X_features)
hidden = 10
w1 = np.random.randn(X_features, hidden)
b1 = np.zeros(hidden)
w2 = np.random.randn(hidden, 1)
b2 = np.zeros(1)
```

```python
# 定位损失函数
def MAE_loss(y,y_hat):
    return np.mean(np.abs(y_hat - y))
# 定义Relu函数
def Relu(x):
    return np.maximum(x,0)
# 学习率
learning_rate=1e-6
# 定义线性回归函数
def Linear(X, W1, b1):
    return X.dot(W1)+b1.reshape((1,W1.shape[1]))
```

```python
# 迭代

for i in range(5000):
    l1=Linear(X,w1,b1)
    r1=Relu(l1)
    y_pred=Linear(r1,w2,b2)
#     计算loss
    loss=np.square(y_pred-Y).sum()
    print('Loss:',loss)
#     反向传播，算梯度
    grad_y_pred=Linear(l1,w2,b2)
    grad_w2=r1.T.dot(grad_y_pred)
    grad_b2=grad_y_pred.mean(axis=0)
    grad_r1=grad_y_pred.dot(w2.T)
    grad_R1=grad_r1.copy()
    grad_R1[r1<0]=0
    grad_w1=X.T.dot(grad_R1)
    grad_b1=grad_R1.mean(axis=0)
#     更细权重
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2
    b1 -= learning_rate * grad_b1
    b2 -= learning_rate * grad_b2
print('w1:{} \n w2:{}'.format(w1, w2))
```
